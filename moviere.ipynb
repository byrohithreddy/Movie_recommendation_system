```python
import pandas as pd
import numpy as np
```

Reading the CSV file

```python
movie = pd.read_csv("dataset.csv")
movie.head()
```

```python
movie.columns
```

```python
movie.info()
```

```python
movie["tags"] = movie["genre"]+movie["overview"]
movie.info()
```

```python
movie = movie.drop(columns = ['genre','overview'])
```

```python
movie.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words="english")
cv
```

```python
vec = cv.fit_transform(movie["tags"].values.astype("U")).toarray()
```

```python
vec
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
movie[movie["title"]=="The Shawshank Redemption"]
```

```python
dist = sorted(list(enumerate(sim[0])),reverse=True,key=lambda vec:vec[1])
```

```python
for i in dist[0:5]:
    print(movie.iloc[i[0]].title)
```

```python
def recommend(movies):
    index = movie[movie["title"]==movies].index[0]
    distance = sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
    for i in distance[0:5]:
        print(movie.iloc[i[0]].title)
```

```python
recommend(movies="Iron Man")
```
